```python
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

def add_months(start_date, months):
    return start_date + relativedelta(months=months)

def generate_roadmap():
    """Generates a roadmap for a student based on their grade, preparation level, and start and end dates."""
    roadmap_tasks = []

    # Get the student's grade and preparation level
    grade = input("Which grade are you currently in? (11th/12th): ").lower()
    preparation_level = input("Have you started your preparation? (yes/no): ")

    start_date_str = input("Enter the start date of your preparation (YYYY-MM-DD): ")
    end_date_str = input("Enter the end date of your entrance exam (YYYY-MM-DD): ")

    start_date = datetime.strptime(start_date_str, "%Y-%m-%d")
    end_date = datetime.strptime(end_date_str, "%Y-%m-%d")



    if grade == "11th":
        if preparation_level.lower() == "yes":
            # Additional tasks for 11th grade and preparation started
            preparation_rating = input("Rate your preparation? (beginner/average/advanced): ")

            # Additional tasks based on preparation level
            if preparation_rating == "beginner":
                roadmap_tasks.append("Complete 11th syllabus :{}".format(add_months(start_date, 8).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt mock test by :{}:".format((start_date + timedelta(days=254)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Complete 12th syllabus by: {}".format(add_months(start_date, 17).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt 12th syllabus mock test by: {}".format((start_date + timedelta(days=560)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Previous year question paper: {}:".format((end_date - timedelta(days=60)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Revision by {}".format((end_date - timedelta(days=40)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Full syllabus test by {}".format((end_date - timedelta(days=10)).strftime("%Y-%m-%d")))
            elif preparation_rating == "average":
                roadmap_tasks.append("Complete 11th syllabus from HC verma by :{}".format(add_months(start_date, 5).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt mock tests by :{}:".format((start_date + timedelta(days=161)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Complete 12th syllabus from HC Verma by: {}".format(add_months(start_date, 11).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt 12th syllabus mock tests by: {}".format((start_date + timedelta(days=338)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt previous year papers by: {}".format((end_date - timedelta(days=90)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Start Revision by {}".format((end_date - timedelta(days=45)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Full syllabus test by {}".format((end_date - timedelta(days=20)).strftime("%Y-%m-%d")))
            else:
                roadmap_tasks.append("Make notes syllabus from NCERT :{}:".format((start_date + timedelta(days=120)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Study HC Verma, IE Erovdo etc : {}".format(add_months(start_date, 6).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt  mock tests by: {}".format((start_date + timedelta(days=213)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Revise your notes: {}".format((end_date - timedelta(days=120)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Previous question papers by {}".format((end_date - timedelta(days=60)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Full syllabus revision {}".format((end_date - timedelta(days=30)).strftime("%Y-%m-%d")))


        elif preparation_level.lower() == "no":
            # Additional tasks for 11th grade and preparation not started
            roadmap_tasks.append("Complete 11th syllabus by :{}".format(add_months(start_date, 8).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Attempt mock test by :{}:".format((start_date + timedelta(days=254)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Complete 12th syllabus by: {}".format(add_months(start_date, 17).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Attempt 12th syllabus mock test by: {}".format((start_date + timedelta(days=560)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Revise 11th syllabus by: {}:".format((end_date - timedelta(days=60)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Previous question papers:{}".format((end_date - timedelta(days=40)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Full syllabus revision {}".format((end_date - timedelta(days=10)).strftime("%Y-%m-%d")))
        else:
            print("Invalid response. Please enter 'yes' or 'no'.")

    elif grade == "12th":
        if preparation_level.lower() == "yes":
            # Additional tasks for 12th grade and preparation started
            preparation_rating = input("Rate your preparation? (beginner/average/advanced): ")

            # Additional tasks based on preparation level
            if preparation_rating == "beginner":
                roadmap_tasks.append("Complete 12th syllabus by :{}".format(add_months(start_date, 4).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Complete 11th syllabus by: {}".format(add_months(start_date, 8).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt full syllabus mock test by: {}".format(add_months(start_date, 9).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Start Revision by {}".format(add_months(start_date, 11).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Full syllabus test by {}".format((end_date - timedelta(days=10)).strftime("%Y-%m-%d")))
            elif preparation_rating == "average":
                roadmap_tasks.append("Complete 12th syllabus by :{}".format(add_months(start_date, 3).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Attempt 12th syllabus mock test by: {}".format((start_date + timedelta(days=103)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Complete 11th syllabus by: {}".format(add_months(start_date, 6).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Study HC verma by: {}".format(add_months(start_date, 9).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Start Revision by {}".format(add_months(start_date, 11).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Full syllabus test by {}".format((end_date - timedelta(days=10)).strftime("%Y-%m-%d")))
            elif preparation_rating == "advanced":
                roadmap_tasks.append("Study HC verma,IE Erodov by: {}".format((start_date + timedelta(days=122)).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Make quick notes by :{}".format(add_months(start_date, 5).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Give full syllabus mock test: {}".format(add_months(start_date, 6).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Start Revision by: {}".format(add_months(start_date, 10).strftime("%Y-%m-%d")))
                roadmap_tasks.append("Solve previous years papers: {}".format((start_date - timedelta(days=30)).strftime("%Y-%m-%d")))
            else:
                print("Invalid preparation rating. Please enter 'beginner', 'average', or 'advanced'.")

        elif preparation_level.lower() == "no":
            # Additional tasks for 12th grade and preparation not started
            roadmap_tasks.append("Complete most commomnly asked topics by :{}".format(add_months(start_date, 4).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Give full syllabus by: {}".format((start_date + timedelta(days=132)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Study easy topics: {}".format(add_months(start_date, 7).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Study moderate topics: {}".format(add_months(start_date, 9).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Solve previous year papers: {}".format((end_date - timedelta(days=60)).strftime("%Y-%m-%d")))
            roadmap_tasks.append("Revise your notes:{}".format((end_date - timedelta(days=7)).strftime("%Y-%m-%d")))
        else:
            print("Invalid response. Please enter 'yes' or 'no'.")

    else:
        print("Invalid grade. Please enter '11th' or '12th'.")

    # Create the roadmap
    roadmap = "\n".join([
        "Start Date:", start_date.strftime("%Y-%m-%d"),
        "End Date:", end_date.strftime("%Y-%m-%d"),
        "",
        "Grade:", grade,
        "Preparation Level:", preparation_rating,
        "",
        "Roadmap:",
    ] + roadmap_tasks)

    return roadmap

# Generate a roadmap for the student
roadmap = generate_roadmap()

# Display the roadmap to the student
print("Your Study Roadmap:")
print(roadmap)

```
